# RAG

## Why RAG?

- AI models can only generate responces based on the data they have been trined on 
- `Gen AI models simply don't know that they don't know`
    - This leads to hallucinations, bias 
- **RAG**: Retrives relevant data from external sources in real time and uses this data to generate more accurate and contextually relevant responses
- It can be seamlessly applied to any type of data => text, images, or audio
- Three main RAG congigurations
    1. naive RAG
        - Does not involve complex data embedding and indexing 
        - It can be efficient to access reasonable amounts of data through keywords
    2. advanced RAG
        - Vector search and and Index based retrieval 
        - It can process multiple data types, as well as multimodel data, which can be structured or unstructured 
    3. modular RAG
        - any scenario that involves naive RAG, advanced RAG, ML and algorithmic projects


### RAG Vs Fine-tuning

- If we accumulate too much data in RAG datasets, the system may become too cumbersome to manage 
- We cannot finetune a model with dynamic, ever changing data such as daily weather forecasts, stock market values..
- The decision of whether RAG or finetuene relies on the propotion of parametric vs non-parametric information 
    - Parametric:
        - Generative AI modelsparameters learned through training data 
        - Models knowladge is stored in these learned weights and biases 
        - The model remebers what it learned from the data , but the data itself is not stored explicitly 
    - Non parametric:
        - Storing data that can be accessed directly 
        - This means that the data remains available and can be queried whenever needed. 
- RAG can be used to improve a models overall efficiency, together with finetunning, which serves as method to enhance the performance of both the retreval and generation components within the RAG framework
