# RAG
Ref: *RAG-Driven Generative AI* - Denis Rothman, ISBN: 978-1-83620-091-8


## Why RAG?

- AI models can only generate responses based on the data they have been trained on.
- GenAI models simply do not know what they do not know.
  - This leads to hallucinations and bias.
- **RAG** retrieves relevant data from external sources in real time and uses this data to generate more accurate and contextually relevant responses.
- It can be applied to any type of data: text, images, or audio.
- Three main RAG configurations:
  1. Naive RAG
     - Does not involve complex data embedding and indexing.
     - It can efficiently access reasonable amounts of data through keywords.
  2. Advanced RAG
     - Uses vector search and index-based retrieval.
     - It can process multiple data types, including multimodal data, whether structured or unstructured.
  3. Modular RAG
     - Covers scenarios that involve naive RAG, advanced RAG, ML, and algorithmic projects.

### RAG vs Fine-tuning

- If we accumulate too much data in RAG datasets, the system may become too cumbersome to manage.
- We cannot fine-tune a model with dynamic, ever-changing data such as daily weather forecasts or stock market values.
- The decision to use RAG or fine-tuning depends on the proportion of parametric vs. non-parametric information.
  - Parametric:
    - Generative AI model parameters learned through training data.
    - Model knowledge is stored in these learned weights and biases.
    - The model remembers what it learned from the data, but the data itself is not stored explicitly.
  - Non-parametric:
    - Data is stored so it can be accessed directly.
    - This means the data remains available and can be queried whenever needed.
- RAG can be used together with fine-tuning to improve overall model efficiency and enhance both retrieval and generation within the RAG framework.

### The RAG ecosystem

*Before choosing a RAG framework, we need to evaluate the proportion of parametric and non-parametric knowledge in the ecosystem we are implementing.*

#### The Generative RAG ecosystem

##### Retriever (D)

- Handles data collection, processing, storage, and retrieval.
- The starting point of a RAG ecosystem is a data ingestion process, whose first step is data collection.

**Collect (D1)**
- Data can be anything.
- A large proportion of data is unstructured and found in unpredictable, complex forms.

**Process (D2)**
- Collected data objects are transformed to create uniform feature representations.
- Data can be chunked, embedded, and indexed.

**Storage (D3)**
- Instead of keeping data in static spreadsheets and files, we turn it into a dynamic, searchable system.
- Examples: Deep Lake, Pinecone, Chroma.

**Retrieval Query (D4)**
- The retrieval process is triggered by user input or automated input (G1).
- To retrieve data quickly, we load it into vector stores and datasets after transforming it into a suitable format.
- Using a combination of keyword search, embeddings, and indexing, we can retrieve data efficiently.
- Cosine similarity is commonly used.

##### The Generator (G)

- User input (G1), automated or human, interacts with the retrieval query (D4) to augment input before sending it to the generative model.

**Input (G1)**
- Batch of automated tasks or human prompts through a user interface (UI).

**Augmented Input with HF (G2)**
- Human feedback can make a RAG ecosystem more adaptable and provide stronger control over data retrieval and generative AI inputs.

**Prompt Engineering (G3)**
- Both the retriever (D) and the generator (G) rely heavily on prompt engineering to prepare standard and augmented messages for the generative AI model to process.

**Generation and Output (G4)**
- The choice of a generative AI model depends on project goals.
- The prompt must meet each model's specifications.

#### Limitations of frontier models
1. Specialized domains
   - Most models are not at PhD-level expertise yet, though they are improving quickly.
2. Recent events
   - Knowledge is limited beyond each model's training cutoff date. Code suggestions may also use legacy APIs/models.
3. Confident mistakes
   - Models can make mistakes confidently.
   - They may have blind spots and jump to conclusions, especially in coding tasks.
